```python
import io
import os

import boto3
import pandas as pd
pd.set_option("display.max_rows", 100)
pd.set_option("display.max_columns", 100)
```

```python
S3_ACCESS_KEY_ID=""
S3_SECRET_ACCESS_KEY=""
```

```python
s3 = boto3.resource(
    "s3",
    aws_access_key_id=S3_ACCESS_KEY_ID,
    aws_secret_access_key=S3_SECRET_ACCESS_KEY
)

def read_parquet_from_s3(bucket_name: str, key: str) -> pd.DataFrame:
    return pd.read_parquet(
        io.BytesIO(
            s3.Object(bucket_name=bucket_name, key=key)
            .get()["Body"]
            .read()
        )
    )
```

```python
BUCKET_NAME = "frozen-facts-center-dev"
KEY = "base_games.parquet"

df = read_parquet_from_s3(bucket_name=BUCKET_NAME, key=KEY)
df
```
